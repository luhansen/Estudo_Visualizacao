qual o gênero mais presente nas aulas?

```python
select_query = "select num_score, university_entrance_call, approved_total_credits_semester, total_credits_semester, average_in_semester from students as S inner join student_course_subject as SCS ON S.id = SCS.id_student where university_entrance_call <> 'Nao informado';"

cursor.execute(select_query)
records = DataFrame(cursor.fetchall())
records.columns = cursor.column_names

records['university_entrance_call'] = records['university_entrance_call'].astype(int)
records['average_in_semester'] = records['average_in_semester'].astype(float)

corr = records.corr()
corr
```

```python
# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
langs = records['student_semester'].apply(str)
students = records['COUNT']
ax.bar(langs,students)
plt.xticks(rotation=90)
plt.show()
```
