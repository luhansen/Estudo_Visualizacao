```python
# Connect to data base
import mysql.connector as mysql
from mysql.connector import Error
try:
    db = mysql.connect(
        host = "localhost",
        database='mestrado2NOTA',
        user = "root",
        passwd = "admin"
    )
    
    if db.is_connected():
        cursor = db.cursor(buffered=True)
#         cursor = db.cursor()
except Error as e:
    print("Error while connecting to MySQL", e)
```

```python
from pandas import DataFrame
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import plotly.express as px
```

## Análise Geral

as notas dos alunos tem melhorado durante os semestres?

```python
select_query="select AVG(num_score) as AVG, semester_year from student_course_subject group by semester_year ;"

cursor.execute(select_query)
records = DataFrame(cursor.fetchall())
records.columns = cursor.column_names

records['semester_year'] = records['semester_year'].astype(int)
```

```python
plt.plot(records['semester_year'].tolist(), records['AVG'].tolist())
plt.xlabel('Ano/Semestre')
plt.ylabel('Média das notas')
plt.show()
```

qual a média das notas dos alunos?

```python
select_query="select num_score from student_course_subject;"

cursor.execute(select_query)
records = DataFrame(cursor.fetchall())
records.columns = cursor.column_names
```

```python
records['num_score'] = records['num_score'].astype(float)
boxplot = records.boxplot(column=['num_score'])
```

```python
ax = sns.violinplot(x="num_score", data=records)
plt.show()
```
