```python
from pandas import DataFrame
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import plotly.express as px
```

## Histórico das notas

### Por sexo

```python
select_query = "select average, total_credits_semester, score, CASE WHEN score = 'SS' THEN 5 WHEN score = 'MS' THEN 4 WHEN score = 'MM' THEN 3 WHEN score = 'MI' THEN 2 WHEN score = 'II' THEN 1 ELSE 0 END AS newScore, semester_year, gender from students as S inner join student_course_subject as SCS on S.id = SCS.id_student inner join subjects as Sub on Sub.id = SCS.id_subject where (name = 'COMPUTACAO BASICA' or name = 'INTRODUCAO A CIENCIA DA COMPUTACAO' or name = 'ALGORITMOS E PROGRAMACAO DE COMPUTADORES');"

cursor.execute(select_query)
records = DataFrame(cursor.fetchall())
records.columns = cursor.column_names
```

```python
F = records[records['gender'] == 'F']
M = records[records['gender'] == 'M']


dataF = F['newScore'].values.tolist()
dataIntF = list(map(float, dataF))
dataIntF = [x + 0.25 for x in dataIntF]

dataM = M['newScore'].values.tolist()
dataIntM = list(map(float, dataM))

plt.title('Frequencia no curso', fontsize=20)
plt.xlabel('Nota', fontsize=15)
plt.ylabel('Frequência Absoluta', fontsize=15)
plt.tick_params(labelsize=15)
plt.gcf().set_size_inches(20, 18)
plt.xticks(rotation=90)
plt.hist(dataIntF, rwidth=0.5, color='red', edgecolor='black')
plt.hist(dataIntM, rwidth=0.5, color='blue', edgecolor='blue')
plt.show()
```

```python
records['average'] = records['average'].astype(float)
fig = px.scatter(records, x="average", y="semester_year",
	         size="newScore", color="gender",
                 log_x=True, size_max=40)
fig.show()
```

```python
records['average'] = records['average'].astype(float)
fig = px.scatter(records, x="average", y="newScore",
	         size="total_credits_semester", color="gender",
                 log_x=True, size_max=40)
fig.show()
```

```python
F['semester_year'] = F['semester_year'].astype(int)
M['semester_year'] = M['semester_year'].astype(int)

plt.scatter(F['semester_year'],F['newScore'], s = 10, c = 'red', label = 'F')
plt.scatter(M['semester_year'],M['newScore'], s = 5, c = 'blue', label = 'M')
plt.title('Iris Clusters and Centroids')
plt.xlabel('média')
plt.ylabel('nota')
plt.legend()
```

```python
records['newScore'] = records['newScore'].astype(float)
boxplot = records.boxplot(column=['newScore'], by="gender", figsize=(7, 6))
```

```python
ax = sns.violinplot(x="newScore", y="gender", data=records)
plt.show()
```
