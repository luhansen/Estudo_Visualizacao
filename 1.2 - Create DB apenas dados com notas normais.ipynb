# Criação das tabelas

```python
# TABELA MATERIAS
# Código.da.disciplina	Nome.da.disciplina	Créditos.disciplina
cursor.execute("DROP TABLE IF EXISTS subjects")
cursor.execute("CREATE TABLE subjects (id INT AUTO_INCREMENT PRIMARY KEY, code VARCHAR(255) NOT NULL, name VARCHAR(255) NOT NULL, credits int(10) NOT NULL)")
```

```python
# TABELA ALUNOS
# 	ID.Aluno	Sexo	UF.de.Nascimento	Cotista	
# Tipo.de.Escola	Raça	Periodo.de.Ingresso.na.Unb	Forma.de.Ingresso.na.UnB
cursor.execute("DROP TABLE IF EXISTS students")
cursor.execute("CREATE TABLE students (id VARCHAR(255) NOT NULL PRIMARY KEY, gender VARCHAR(255) NOT NULL, uf_birth VARCHAR(10) NOT NULL, quotaholder VARCHAR(10) NOT NULL, school_type VARCHAR(255) NOT NULL, race VARCHAR(255) NOT NULL, university_entrance_period VARCHAR(255) NOT NULL, entrance_form VARCHAR(255))")
```

```python
# TABELA ALUNOS_CURSO
# ID, ID.Aluno, Curso, Opção, Periodo.de.ingresso.na.opção, Período.de.saída.da.opção, 
# Forma.de.saída.da.opção, Min..Créd..Formatura	
# Créditos.cursados.no.total	Créditos.integralizados.no.total	Créditos.a.integralizar.no.total
cursor.execute("DROP TABLE IF EXISTS student_courses")
cursor.execute("CREATE TABLE student_courses (id INT AUTO_INCREMENT PRIMARY KEY, course VARCHAR(255) NOT NULL, opt VARCHAR(255) NOT NULL, option_entrance_period VARCHAR(255) NOT NULL, leaving_form VARCHAR(255) NOT NULL, leaving_period VARCHAR(255) NOT NULL, minimum_graduation_credit VARCHAR(255) NOT NULL, total_credits VARCHAR(255) NOT NULL, paidin_credits VARCHAR(255) NOT NULL, credits_to_be_paid VARCHAR(255) NOT NULL, id_student VARCHAR(255) NOT NULL, FOREIGN KEY(id_student) REFERENCES students(id) ON DELETE CASCADE)")
```

```python
# TABELA_ALUNOS_MATERIA_CURSO
# Ano.e.Semestre	Média.do.período	Ano.e.Semestre
# Média.do.período}	ID.Aluno	(Código.da.disciplina)	(ID.Aluno.Curso)
cursor.execute("DROP TABLE IF EXISTS student_course_subject")
cursor.execute("CREATE TABLE student_course_subject(id INT AUTO_INCREMENT PRIMARY KEY, score VARCHAR(255) NOT NULL, average VARCHAR(255) NOT NULL, semester_year VARCHAR(255) NOT NULL, total_credits_semester int NOT NULL, id_student VARCHAR(255) NOT NULL, id_student_course INT NOT NULL, id_subject INT NOT NULL, FOREIGN KEY(id_student) REFERENCES students(id) ON DELETE CASCADE, FOREIGN KEY(id_student_course) REFERENCES student_courses(id) ON DELETE CASCADE, FOREIGN KEY(id_subject) REFERENCES subjects(id) ON DELETE CASCADE)")
```

## Student Courses
na tabela 'student_courses' um mesmo aluno pode ter feito dois cursos diferentes, tendo duas instancias na tabela

```python
values_student_courses = all_data[['ID.Aluno', 'Curso', 'Opção', 'Periodo.de.ingresso.na.opção', 'Período.de.saída.da.opção', 'Forma.de.saída.da.opção', 'Min..Créd..Formatura', 'Créditos.cursados.no.total', 'Créditos.integralizados.no.total', 'Créditos.a.integralizar.no.total']]
uniqueValuesStudentCourses = values_student_courses.drop_duplicates()
```

```python
print(uniqueValuesStudentCourses['ID.Aluno'].count())
print(values_student_courses['ID.Aluno'].count())
```

na tabela 'student_courses' um aluno acaba estando ativo em 2 cursos com mesmo nome, mas total de creditos distintos.

```python
def arrumar_curso(row, curso, min_creditos_antigo, min_creditos_novo):
    if((row['Curso'] == curso) & (row['Min..Créd..Formatura'] == min_creditos_antigo)):
        found2 = uniqueValuesStudentCourses[(uniqueValuesStudentCourses['ID.Aluno'] == row['ID.Aluno']) & (uniqueValuesStudentCourses['Curso'] == curso) & (uniqueValuesStudentCourses['Min..Créd..Formatura'] == min_creditos_novo) & (uniqueValuesStudentCourses['Forma.de.saída.da.opção'] == row['Forma.de.saída.da.opção']) & (uniqueValuesStudentCourses['Periodo.de.ingresso.na.opção'] == row['Periodo.de.ingresso.na.opção'])]
        if(len(found2) > 0):
            return 'Mudança de curriculo'
    return row['Forma.de.saída.da.opção']
```

```python
def arrumar_periodo_de_saida(row, ano, creditos):
    if((row['Forma.de.saída.da.opção'] == 'Mudança de curriculo') & (row['Min..Créd..Formatura'] == creditos)):
        return ano
    return row['Período.de.saída.da.opção']
```

```python
uniqueValuesStudentCourses['Forma.de.saída.da.opção'] = uniqueValuesStudentCourses.apply(lambda x: arrumar_curso(x, 'Ciência da Computação', 240, 214), axis=1)
uniqueValuesStudentCourses['Forma.de.saída.da.opção'] = uniqueValuesStudentCourses.apply(lambda x: arrumar_curso(x, 'Ciência da Computação', 242, 240), axis=1)
uniqueValuesStudentCourses['Forma.de.saída.da.opção'] = uniqueValuesStudentCourses.apply(lambda x: arrumar_curso(x, 'Engenharia de Redes de Comunicação', 244, 250), axis=1)
```

```python
uniqueValuesStudentCourses['Período.de.saída.da.opção'] = uniqueValuesStudentCourses.apply(lambda x: arrumar_periodo_de_saida(x, 20051, 244), axis=1)
uniqueValuesStudentCourses['Período.de.saída.da.opção'] = uniqueValuesStudentCourses.apply(lambda x: arrumar_periodo_de_saida(x, 19951, 242), axis=1)
uniqueValuesStudentCourses['Período.de.saída.da.opção'] = uniqueValuesStudentCourses.apply(lambda x: arrumar_periodo_de_saida(x, 20152, 240), axis=1)
```

```python
# Tabela Aluno_curso_materia
for i,row in all_data.iterrows():
#     print(tuple(row))
    select_idcurso = "SELECT id FROM student_courses WHERE id_student = %s AND course = %s AND option_entrance_period = %s AND minimum_graduation_credit = %s"
    cursor.execute(select_idcurso, (row['ID.Aluno'], row['Curso'], row['Periodo.de.ingresso.na.opção'], row['Min..Créd..Formatura']) )

    ## fetching all usernames from the 'cursor' object
    id_student_courses = cursor.fetchall()
    if len(id_student_courses) > 1:
        print('Erro', id_student_courses)
        exit()

    id_student_courses = id_student_courses[0]

    select_idsubject = "SELECT id from subjects WHERE code = %s AND name = %s AND credits = %s"
    cursor.execute(select_idsubject, (row['Código.da.disciplina'], row['Nome.da.disciplina'],  row['Créditos.disciplina']) )
    
    id_subjects = cursor.fetchall()
    if len(id_subjects) > 1:
        print('Erro', id_subjects)
        exit()

    id_subjects = id_subjects[0]

    sql = "INSERT INTO student_course_subject (id_student, id_student_course, id_subject, score, average, semester_year, total_credits_semester) VALUES (%s, %s, %s, %s, %s, %s, %s)"
    cursor.execute(sql, (row['ID.Aluno'], id_student_courses[0], id_subjects[0], row['Menção.na.disciplina'], row['Média.do.período'], row['Ano.e.Semestre'], row['Créditos.cursados.no.semestre.com.aprovação.']))
    db.commit()
```
