## Tirar espaço no final

```python
all_data['Aluno'] = all_data['Aluno'].apply(lambda x: unidecode(x.strip()))
all_data['curso'] = all_data['curso'].apply(lambda x: unidecode(x.strip()))
all_data['forma_ingresso_unb'] = all_data['forma_ingresso_unb'].apply(lambda x: unidecode(x.strip()))
all_data['nome_disciplina'] = all_data['nome_disciplina'].apply(lambda x: unidecode(x.strip()))
```

## Tratamento de NaN value

```python
all_data[all_data['chamada_ingressou_UnB'].isnull()]
```

```python
def generalizar_forma_saida(row):
    if not pd.isnull(row['forma_saida_curso']):
        row['forma_saida_curso'] = row['forma_saida_curso'].strip()
        if((row['forma_saida_curso'] != 'Formatura') & (row['forma_saida_curso'] != 'Falecimento')):
            return 'Saiu'
        return row['forma_saida_curso']
    else:
        return 'Ativo'
```

```python
all_data['forma_saida_curso'] = all_data.apply(lambda x: generalizar_forma_saida(x), axis=1)
```

```python
def endereco_nao_informado(row, column):
    if not pd.isnull(row[column]):
        return row[column]
    else:
        return 'Nao informado'
```

```python
all_data['endereco'] = all_data.apply(lambda x: endereco_nao_informado(x, 'endereco'), axis=1)
all_data['chamada_ingressou_UnB'] = all_data.apply(lambda x: endereco_nao_informado(x, 'chamada_ingressou_UnB'), axis=1)
```

```python
def sem_cota(row):
    if not pd.isnull(row['cota']):
        return row['cota']
    else:
        return 'Nao'
```

```python
all_data['cota'] = all_data.apply(lambda x: sem_cota(x), axis=1)
```

## Tratamento Extra

### periodo de saida -> ativo = 0

```python
def generalizar_periodo_saida(row):
    row['periodo_saida_curso'] = row['periodo_saida_curso'].strip()
    if(row['periodo_saida_curso'] == 'ativo'):
        return 0
    return int(row['periodo_saida_curso'])
```

```python
all_data['periodo_saida_curso'] = all_data.apply(lambda x: generalizar_periodo_saida(x), axis=1)
```

### identificar alunos unicos

```python
values_students = all_data[['Aluno', 'genero', 'nascimento', 'endereco', 'estado_nascimento', 'sistema_cotas', 'cota', 'Escola', 'periodo_ingresso_unb', 'forma_ingresso_unb', 'chamada_ingressou_UnB']]
uniqueValuesStudents = values_students.drop_duplicates()
```

```python
print(uniqueValuesStudents['Aluno'].count())
print(values_students['Aluno'].count())
print(len(all_data['Aluno'].unique().tolist()))
print(len(all_data['Id_pessoa'].unique().tolist()))
```

```python
uniqueValuesStudents.insert(0, 'New_ID', range(1, 1 + len(uniqueValuesStudents)))
```

```python
for index, row in uniqueValuesStudents.iterrows():
    new_id = row['New_ID'];
    selected_data = all_data.loc[(all_data['Aluno'] == row['Aluno']) &
            (all_data['genero'] == row['genero']) &
            (all_data['nascimento'] == row['nascimento']) & 
            (all_data['endereco'] == row['endereco']) & 
            (all_data['estado_nascimento'] == row['estado_nascimento']) &
            (all_data['sistema_cotas'] == row['sistema_cotas']) &
            (all_data['cota'] == row['cota']) &
            (all_data['Escola'] == row['Escola']) &
            (all_data['periodo_ingresso_unb'] == row['periodo_ingresso_unb']) &
            (all_data['forma_ingresso_unb'] == row['forma_ingresso_unb']) &
            (all_data['chamada_ingressou_UnB'] == row['chamada_ingressou_UnB'])]
#             (all_data['IRA'] == row['IRA']) &
#             (all_data['total_creditos_cursados_aluno'] == row['total_creditos_cursados_aluno'])]
    if selected_data.empty:
        print(row, index, new_id)
    for selected_index, selected_row in selected_data.iterrows():
        all_data.loc[selected_index, 'id_student'] = new_id
#         selected_row['id_student'] = new_id
```
