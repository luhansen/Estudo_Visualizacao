```python
# quando o aluno muda de curso, ele costuma vir de qual curso?
select_query = "select id_student, (SELECT group_concat(name order by student_semester asc) as bla from student_course_subject SCS INNER JOIN subjects as S ON S.id=SCS.id_subject WHERE SCS.id_student= t1.id_student) as disc from student_course_subject t1 group by id_student ;"
cursor.execute(select_query)
records = DataFrame(cursor.fetchall())
records.columns = cursor.column_names

s = pd.Series(records['disc'].to_list())
df = s.str.split(',', expand=True)
mask = df.applymap(lambda x: x is None)
cols = df.columns[(mask).any()]
for col in df[cols]:
    df.loc[mask[col], col] = 'Nenhum'
# df = df.drop([1])
df.columns = [x for x in range(1, 43)]
df.index = pd.Series(records['id_student'].to_list()).drop_duplicates()

df2 = df[df[2] != 'Nenhum']
fig = px.treemap(
    df2, path=[1, 2, 3]
)
fig.show()
```
